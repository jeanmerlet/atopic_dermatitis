```python
import pandas as pd
import numpy as np
import umap
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
cm = plt.cm.get_cmap('tab20')
n = 750
colors = []
for i in range(n):
    colors.append(cm.colors[np.random.randint(0, 3)])
x, y = np.random.rand(2, n)
fig, ax = plt.subplots(figsize=(9, 6))
plot = ax.scatter(x, y, c=colors, facecolors='none')
handles = []
for i in range(3):
    artist = Line2D([0], [0], color='white', marker='o', markerfacecolor=cm.colors[i],
                    label='cluster ' + str(i), markersize=9)
    handles.append(artist)
ax.legend(handles=handles, bbox_to_anchor=(1.02,1), loc="upper left")
plt.show()
```
